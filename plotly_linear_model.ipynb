```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
import plotly.express as px

df = px.data.tips()
fig = px.scatter(df, x="total_bill", y="tip", trendline="ols")
fig.show()
```

Resize figure:
* https://stackoverflow.com/questions/332289/how-do-you-change-the-size-of-figures-drawn-with-matplotlib#638443    

```python
x = np.linspace(0, 100, 50)
y = 2*x + 1

fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(x, y)
plt.show()
```

```python
fig = px.scatter(df, x=x, y=y, trendline="ols")
fig.show()
```

Adding noise/randomness to the data:  
https://stackoverflow.com/questions/14058340/adding-noise-to-a-signal-in-python#14058425

```python
noise = np.random.normal(0, 7, y.shape)
y_2 = y + noise
```

```python
y_2
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
plt.plot(x, y_2)
plt.show()
```

```python
fig = px.scatter(df, x=x, y=y_2, trendline="ols")
fig.show()
```
